```python
import pandas as pd

path = "/Users/mkapilev/Documents/Insight/Project_databases/survey_data/csv/demographic-snapshot-2015-16-to-2019-20-(public).csv"
enrollmentDF = pd.read_csv(path)
enrollmentDF = enrollmentDF[['DBN', 'School Name', 'Year', 'Total Enrollment', 'Economic Need Index', 'Enrollment Change']]
enrollmentDF = enrollmentDF[(enrollmentDF['Year'] != '2019-20')] # filter out last year 2020 - we don't have data for it
enrollmentDF
```

```python
import pandas as pd

studentSurveyDF = pd.read_excel('/Users/mkapilev/Documents/Insight/Project_databases/survey_data/xlsx/2019-public-data-file_student.xlsx', sheet_name='Student % Pos & Neg %')
studentSurveyDF['Year']='2018-19'
studentSurveyDF

unnamedCols_S = []

for col in studentSurveyDF.columns: 
    if col.startswith("Unnamed:"):
        unnamedCols_S.append(col)

studentSurveyDF = studentSurveyDF.drop(columns=unnamedCols_S)
studentSurveyDF.drop([0,1], inplace=True)
studentSurveyDF
```

```python
joinedDF_student = enrollmentDF.merge(studentSurveyDF, how='inner', left_on=['DBN', 'Year'], right_on=['DBN', 'Year'])

for col in joinedDF_student.columns:
    print(col)
```
