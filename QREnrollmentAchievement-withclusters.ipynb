```python
a = achievementMiddleDF[['DBN', 'BN','Student Achievement - Section Score']]
b = achievementHighDF[['DBN', 'BN','Student Achievement - Section Score']]
achievementDF = pd.concat([a, b])
achievementDF.drop([0], inplace=True)

achievementDF
```

```python
def isfloat(value):
    try:
        float(value)
        return True
    except ValueError:
        return False

def valid(row):
    return isfloat(row['Student Achievement - Section Score'])

def normalizeScore(row):
    return float(row['Student Achievement - Section Score'])/5

achievementDF1 = achievementDF[achievementDF.apply(valid, axis=1)]
achievementDF1['Student Achievement - Section Score Normalized'] = achievementDF1.apply(normalizeScore, axis=1)
achievementDF1
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

def run_regression(cluster_id):
    myclusterDF = joinedDF1[joinedDF1['Cluster'] == cluster_id]
    X = myclusterDF[["Indicator_1.1","Indicator_1.2","Indicator_1.3","Indicator_2.2", "Indicator_3.1","Indicator_4.1","Indicator_4.2", "Indicator_5.1"]] 
    y = myclusterDF['Student Achievement - Section Score Normalized']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # actually learn the model
    model = linear_model.LinearRegression()
    #model = linear_model.ElasticNet(alpha=1.0, l1_ratio=0.9)
    model.fit(X_train, y_train)
    #TODO actual stuff and all
    y_pred = model.predict(X_test)

    # The coefficients
    print('Coefficients: \n', model.coef_)

    # The mean squared error
    print('Mean squared error: %.2f'% mean_squared_error(y_test, y_pred))

    # The coefficient of determination: 1 is perfect prediction
    print('Coefficient of determination: %.2f'% r2_score(y_test, y_pred))

    print(f'R^2 score: {model.score(X_test, y_test)}')
    
num_clusters = 5    
    
for i in range(num_clusters):
    run_regression(i)
```
