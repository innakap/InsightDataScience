```python
import pandas as pd

path = "/Users/mkapilev/Documents/Insight/Project_databases/survey_data/csv/demographic-snapshot-2015-16-to-2019-20-(public).csv"
enrollmentDF = pd.read_csv(path)
enrollmentDF = enrollmentDF[['DBN', 'School Name', 'Year', 'Total Enrollment', 'Economic Need Index', 'Enrollment Change']]
enrollmentDF = enrollmentDF[(enrollmentDF['Year'] != '2019-20')] # filter out last year 2020 - we don't have data for it
enrollmentDF
```

```python
import pandas as pd

TotalSurveyDF = pd.read_excel('/Users/mkapilev/Documents/Insight/Project_databases/survey_data/xlsx/2019-public-data-file_parent.xlsx', sheet_name='Total')
TotalSurveyDF['Year']='2018-19'
TotalSurveyDF

```

```python
TotalSurveyDF1 = TotalSurveyDF.drop(columns=['Total Parent \nResponse Rate', 'Total Teacher Response Rate'])
TotalSurveyDF.drop([0,1], inplace=True)
TotalSurveyDF1.head(100)
```

```python
TotalSurveyDF1.drop([0,1], inplace=True)
TotalSurveyDF1.head(100)
```

```python
# https://stackoverflow.com/questions/11418192/pandas-complex-filter-on-rows-of-dataframe

def high_school_filter(row):
    return 'HIGH SCHOOL' in row['School Name']

TotalSurveyDF2=TotalSurveyDF1[TotalSurveyDF1.apply(high_school_filter, axis=1)]
TotalSurveyDF2
```

```python
# joinedDF = enrollmentDF.merge(TotalSurveyDF1, how='inner', left_on=['DBN', 'Year'], right_on=['DBN', 'Year'])
joinedDF2 = enrollmentDF.merge(TotalSurveyDF2, how='inner', left_on=['DBN', 'Year'], right_on=['DBN', 'Year'])
joinedDF2

# for col in joinedDF.columns:
#     print(col)
```

```python
naCounts = joinedDF2.isna().sum()
naCounts
```

```python
missingValueColumns = naCounts[naCounts >= 200].keys().values
missingValueColumns
```

```python
joinedDF2 = joinedDF2.drop(columns=missingValueColumns) # remove all those N/A columns
joinedDF2
```

```python
print(joinedDF2.shape)
joinedDF2 = joinedDF2.dropna(how='any', axis=0)
# print(joinedDF2.dropna(how='any', axis=0).shape)

# now remove any rows with N/A. number of rows decreased
#  


print(joinedDF2.shape)
```

```python
import re

# def to_percent(num):
#     return 1 - num / 100

# feature_columns = []

# now we're actually gonna create our X feature matrix, using just the question responses as features

# for col in joinedDF.columns:
#     if re.match(r"[0-9]+[a-z]\.", col):
#         feature_columns.append(col)

X = joinedDF2[["Collaborative Teachers Score","Effective School Leadership Score","Rigorous Instruction Score","Strong Family-Community Ties Score", "Trust Score"]] # select just the columns we want as features
# X = X.apply(to_percent) # divide everything by 100. smaller values are usually better for learning.
X
```

```python
y = joinedDF2['Enrollment Change']
y
```

```python
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# actually learn the model
model = linear_model.LinearRegression()
#model = linear_model.ElasticNet(alpha=1.0, l1_ratio=0.9)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

# The coefficients
print('Coefficients: \n', model.coef_)

# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_test, y_pred))

# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_test, y_pred))

print(f'R^2 score: {model.score(X_test, y_test)}')
```

```python
for pred, actual in zip(y_pred, y_test):
    print(f"predicted={pred}\tactual={actual}")
```
