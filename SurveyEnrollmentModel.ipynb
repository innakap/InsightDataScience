```python
import pandas as pd

path = "/Users/mkapilev/Documents/Insight/Project_databases/survey_data/csv/demographic-snapshot-2015-16-to-2019-20-(public).csv"
enrollmentDF = pd.read_csv(path)
enrollmentDF = enrollmentDF[['DBN', 'School Name', 'Year', 'Total Enrollment', 'Economic Need Index', 'Enrollment Change']]
enrollmentDF = enrollmentDF[(enrollmentDF['Year'] != '2019-20')] # filter out last year 2020 - we don't have data for it
enrollmentDF
```

```python
import pandas as pd

parentSurveyDF = pd.read_excel('/Users/mkapilev/Documents/Insight/Project_databases/survey_data/xlsx/2019-public-data-file_parent.xlsx', sheet_name='Parent % Pos & Neg %')
parentSurveyDF['Year']='2018-19'
#parentSurveyDF

unnamedCols = []

for col in parentSurveyDF.columns: 
    if col.startswith("Unnamed:"):
        unnamedCols.append(col)

parentSurveyDF = parentSurveyDF.drop(columns=unnamedCols)
parentSurveyDF.drop([0,1], inplace=True)
parentSurveyDF.head(100)
```

```python
joinedDF = enrollmentDF.merge(parentSurveyDF, how='inner', left_on=['DBN', 'Year'], right_on=['DBN', 'Year'])

for col in joinedDF.columns:
    print(col)
```

```python
naCounts = joinedDF.isna().sum()
naCounts
```

```python
missingValueColumns = naCounts[naCounts >= 1000].keys().values
missingValueColumns
```

```python
joinedDF = joinedDF.drop(columns=missingValueColumns) # remove all those N/A columns
joinedDF
```

```python
print(joinedDF.shape)

# now remove any rows with N/A. you will see the number of rows decreased
joinedDF = joinedDF.dropna(how='any', axis=0) 

print(joinedDF.shape)
```

```python
import re

def to_percent(num):
    return 1 - num / 100

feature_columns = []

# now we're actually gonna create our X feature matrix, using just the question responses as features

for col in joinedDF.columns:
    if re.match(r"[0-9]+[a-z]\.", col):
        feature_columns.append(col)

X = joinedDF[feature_columns] # select just the columns we want as features
X = X.apply(to_percent) # divide everything by 100. smaller values are usually better for learning.
X
```

```python
y = joinedDF['Enrollment Change']
y
```

```python
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# actually learn the model
model = linear_model.LinearRegression()
#model = linear_model.ElasticNet(alpha=1.0, l1_ratio=0.9)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

# The coefficients
print('Coefficients: \n', model.coef_)

# The mean squared error
print('Mean squared error: %.2f'% mean_squared_error(y_test, y_pred))

# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'% r2_score(y_test, y_pred))

print(f'R^2 score: {model.score(X_test, y_test)}')
```

```python
for pred, actual in zip(y_pred, y_test):
    print(f"predicted={pred}\tactual={actual}")
```
