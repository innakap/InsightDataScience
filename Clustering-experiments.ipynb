```python
a = achievementMiddleDF[['DBN', 'Student Achievement - Section Score']]
b = achievementHighDF[['DBN', 'Student Achievement - Section Score']]
achievementDF = pd.concat([a, b])
achievementDF
```

```python
joinedDF = demographicDF[['DBN', 'Economic Need Index']].merge(achievementDF, how='inner', left_on=['DBN'], right_on=['DBN'])
joinedDF
```

```python
def isfloat(value):
    try:
        float(value)
        return True
    except ValueError:
        return False

def valid(row):
    return isfloat(row['Economic Need Index']) and isfloat(row['Student Achievement - Section Score'])

def parseScore(row):
    return float(row['Student Achievement - Section Score'])

def normalizeScore(row):
    return float(row['Student Achievement - Section Score']) / 5

joinedDF = joinedDF[joinedDF.apply(valid, axis=1)]
joinedDF['Student Achievement - Section Score'] = joinedDF.apply(parseScore, axis=1)
joinedDF['Student Achievement - Section Score Normalized'] = joinedDF.apply(normalizeScore, axis=1)
joinedDF
```

```python
joinedDF.dtypes
```

```python
from sklearn.cluster import KMeans

X = joinedDF[['Economic Need Index', 'Student Achievement - Section Score Normalized']]

km = KMeans(
    n_clusters=5, init='random',
    n_init=10, max_iter=300, 
    tol=1e-04, random_state=0
)
y_km = km.fit_predict(X)

km.cluster_centers_
```

```python
clusterDF = joinedDF
clusterDF['Cluster'] = y_km
clusterDF.dtypes
```

```python
pd.Series(y_km).value_counts()
```

```python
#Nearest Neighbors
from sklearn.neighbors import KDTree
import numpy as np
kdt = KDTree(X, leaf_size=40, metric='euclidean')

# find 10 closest schools given the target school
school = np.array([[0.869011, 0.972]])
kdt.query(school, k=10, return_distance=True)
```

```python
#Nearest Neighbors
from sklearn.neighbors import NearestNeighbors
import numpy as np

school = np.array([[0.869011, 0.972]])
nbrs = NearestNeighbors(n_neighbors=10, algorithm='ball_tree').fit(X)
distances, indices = nbrs.kneighbors(school)
distances, indices
```

```python
def run_correlation(cluster_id):
    myclusterDF = joinedDF2[joinedDF2['Cluster'] == cluster_id]
    return myclusterDF.corr()['Student Achievement - Section Score']

num_clusters = 5

correlations = []

for i in range(num_clusters):
    series = run_correlation(i)
    series.set_value('Cluster', i)
    correlations.append(series)
    
pd.concat(correlations, axis=1)
```

```python
import numpy as np
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

def run_regression(cluster_id):
    myclusterDF = joinedDF2[joinedDF2['Cluster'] == cluster_id]
    X = myclusterDF[["Collaborative Teachers Score","Effective School Leadership Score","Rigorous Instruction Score","Strong Family-Community Ties Score", "Trust Score"]] 
    y = myclusterDF['Student Achievement - Section Score']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # actually learn the model
    model = linear_model.LinearRegression()
    #model = linear_model.ElasticNet(alpha=1.0, l1_ratio=0.9)
    model.fit(X_train, y_train)
    #TODO actual stuff and all
    y_pred = model.predict(X_test)

    # The coefficients
    print('Coefficients: \n', model.coef_)

    # The mean squared error
    print('Mean squared error: %.2f'% mean_squared_error(y_test, y_pred))

    # The coefficient of determination: 1 is perfect prediction
    print('Coefficient of determination: %.2f'% r2_score(y_test, y_pred))

    print(f'R^2 score: {model.score(X_test, y_test)}')
    
num_clusters = 5    
    
for i in range(num_clusters):
    run_regression(i)
```
